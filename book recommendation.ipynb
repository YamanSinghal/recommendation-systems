```python
books = pd.read_csv("/kaggle/input/bookcrossing/BX-Books.csv",sep = ";",error_bad_lines = False,encoding = "latin-1")
users = pd.read_csv("/kaggle/input/bookcrossing/BX-Users.csv",sep = ";",error_bad_lines = False,encoding = "latin-1")
ratings = pd.read_csv("/kaggle/input/bookcrossing/BX-Book-Ratings.csv",sep = ";",error_bad_lines = False,encoding = "latin-1")
```

```python
books = books.iloc[:,0:4]
books.head()
```

```python
books["Book-Title"].value_counts()
```

```python
users.fillna(users["Age"].mean(),inplace = True)
users.head()
```

```python
ratings.head()
```

```python
book_rating = books.merge(ratings,on = "ISBN")
book_rating 
#merging books with ratings
```

```python
book_rating.shape
```

```python
b = book_rating["Book-Title"].value_counts() > 50
#getting books which received ratings by more than 100 users
```

```python
b[b].shape
#the number of books which received ratings by more than 100 users
```

```python
a = b[b].index
#getting names of books 
```

```python
book_rating = book_rating[book_rating["Book-Title"].isin(a)]
```

```python
book_rating.shape
```

```python
book_rating
```

```python
u = book_rating["User-ID"].value_counts() > 200
u[u]
```

```python
u[u].shape
```

```python
w = u[u].index
w
```

```python
book_rating = book_rating[book_rating["User-ID"].isin(w)]
book_rating
```

```python
book_rating.shape
```

```python
book_rating.drop_duplicates(["User-ID","Book-Title"],inplace = True)
```

```python
book_rating.shape
```

```python
book_pivot = book_rating.pivot(columns = "User-ID",index = "Book-Title",values = "Book-Rating")
```

```python
book_pivot
```

```python
book_pivot.fillna(0,inplace = True)
book_pivot
```

```python
from scipy.sparse import csr_matrix
sparse = csr_matrix(book_pivot)
```

```python
from sklearn.neighbors import NearestNeighbors
nn = NearestNeighbors(algorithm = "brute")
nn.fit(sparse)
```

```python
dist,recommendation = nn.kneighbors(book_pivot.iloc[766,:].values.reshape(1,-1),n_neighbors = 6)
```
